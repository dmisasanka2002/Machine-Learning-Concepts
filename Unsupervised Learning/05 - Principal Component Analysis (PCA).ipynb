# **Principal Component Analysis (PCA)**

## **Introduction**

Principal Component Analysis (PCA) is an *unsupervised dimensionality reduction technique* widely used in machine learning and data preprocessing. PCA transforms high-dimensional data into a lower-dimensional space by finding the directions (principal components) that maximize variance while minimizing information loss.


##
---

## **Key Concepts**

1. **Dimensionality Reduction**: PCA reduces the number of features in a dataset while retaining as much variability as possible.
2. **Principal Components**: Orthogonal vectors that represent the directions of maximum variance in the data.
3. **Linear Transformation**: PCA is a linear transformation technique that projects data into a new coordinate system.


##
---

## **Working Method**

### **Steps in PCA**

1. **Standardize the Data**:
   Center the data by subtracting the mean and scaling by the standard deviation.

2. **Compute the Covariance Matrix**:
   The covariance matrix represents the relationship between features:
   $$
   \Sigma = \frac{1}{n} \sum_{i=1}^{n} (X_i - \mu)(X_i - \mu)^T
   $$

3. **Calculate Eigenvalues and Eigenvectors**:
   - **Eigenvalues**: Measure the variance explained by each principal component.
   - **Eigenvectors**: Represent the directions of the principal components.

4. **Sort Components by Variance**:
   Select the top \(k\) principal components that explain the majority of the variance.

5. **Transform Data**:
   Project the original data onto the principal components:
   $$
   X_{PCA} = X \cdot W
   $$
   Where \( W \) is the matrix of top \(k\) eigenvectors.


##
---

## **Mathematical Concepts**

### **1. Covariance Matrix**

The covariance matrix measures the pairwise relationships between features:
$$
\text{Cov}(X, Y) = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{n - 1}
$$

### **2. Eigenvalues and Eigenvectors**

- **Eigenvalues (\( $\lambda$ \))**:
  Represent the amount of variance explained by each principal component.
- **Eigenvectors (\( v \))**:
  Indicate the direction of the principal components.

The relationship is given by:
$$
\Sigma v = \lambda v
$$

### **3. Explained Variance**

The proportion of variance explained by each principal component:
$$
\text{Explained Variance Ratio} = \frac{\lambda_i}{\sum_{j=1}^{p} \lambda_j}
$$


##
---

## **Types of PCA**

### **1. Standard PCA**
Uses the entire dataset without incorporating additional information.

### **2. Kernel PCA**
Extends PCA to nonlinear data by applying the kernel trick, transforming data into higher-dimensional space before applying PCA.

### **3. Sparse PCA**
Applies sparsity constraints to the principal components, often used for feature selection.

## **Python Implementation**

### **1. Standard PCA**

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 5)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualize the results
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='k', s=50)
plt.title("PCA - 2 Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# Explained variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

````python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 5)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualize the results
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='k', s=50)
plt.title("PCA - 2 Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# Explained variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
````

### **2. Kernel PCA**

```python
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_circles

# Generate nonlinear data
X, y = make_circles(n_samples=300, factor=0.3, noise=0.05)

# Apply Kernel PCA with RBF kernel
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)

# Visualize the results
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("Kernel PCA - RBF Kernel")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

````python
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_circles

# Generate nonlinear data
X, y = make_circles(n_samples=300, factor=0.3, noise=0.05)

# Apply Kernel PCA with RBF kernel
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)

# Visualize the results
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title("Kernel PCA - RBF Kernel")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
````

````python
from sklearn.decomposition import SparsePCA

# Apply Sparse PCA
spca = SparsePCA(n_components=2, random_state=42)
X_spca = spca.fit_transform(X_scaled)

# Visualize the results
plt.scatter(X_spca[:, 0], X_spca[:, 1], c='orange', edgecolor='k', s=50)
plt.title("Sparse PCA - 2 Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
````

## **Advantages & Limitations**

### Advantages

 - Reduces overfitting by eliminating less important features.

 - Improves computational efficiency for high-dimensional data.

 - Provides insights into the structure of the data.

### Limitations

 - Assumes linearity in relationships between features.

 - Sensitive to scaling and outliers.

 - Principal components may lack interpretability.

##
---

## **Applications**

- Data visualization in lower dimensions (2D or 3D).
- Noise reduction.
- Feature extraction for supervised learning algorithms.
- Speeding up computations in high-dimensional datasets.

##
---

## **Conclusion**

PCA is a powerful dimensionality reduction tool that simplifies complex data into interpretable components while preserving the maximum variance. Its extensions, such as Kernel PCA and Sparse PCA, further enhance its capability to handle nonlinear and sparse datasets.
