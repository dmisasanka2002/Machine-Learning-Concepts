```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
data = np.random.rand(100, 2)
X = 20 * data[:,0]
y = data[:, 1]
X.shape
```

```python
data = np.array([X,y]).reshape(100, 2)
data.shape
```

```python
# Plot data 
plt.scatter(X, y)
```

```python
# Standardize the data
scaler = StandardScaler()
scaled = scaler.fit_transform(data)
scaled
```

```python
# Plot data 
plt.scatter(scaled[:,0], scaled[:,1])
```

- **n_components**: must be between 0 and min(n_samples, n_features)

```python
# Apply PCA
pca = PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(scaled)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1])
```

- Convert 2D data into 1D data

```python
# Apply PCA
pca = PCA(n_components=1) # n_components=1 : for 1D
```

```python
pca_data = pca.fit_transform(scaled)
pca_data.shape
```

```python
pca.explained_variance_ratio_
```

```python
data_1D = pca.transform(scaled)
data_1D.shape
```

```python
scaled.shape
```

```python
data_1D = pca.inverse_transform(data_1D)
data_1D[0:5, :]
```

```python
plt.scatter(scaled[:,0], scaled[:,1])
plt.scatter(data_1D[:,0], data_1D[:,1], color="g")
plt.show()
```

### PCA for visualization

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
digits.data.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.matshow(digits.images[333])
plt.show() 
```

```python
digits.target[333]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # get 2D data
new_data = pca.fit_transform(digits.data)
print(digits.data.shape)
print(new_data.shape)
```

```python
plt.scatter(new_data[:, 0], new_data[:, 1], c=digits.target)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.colorbar()
plt.show()
```
