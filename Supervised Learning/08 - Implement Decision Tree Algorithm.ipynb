```python
import pandas as pd
```

## **Decision Tree Algorithm for Classification Problem**

Below is a Python example using *Decision Tree* for a classification task on the Iris dataset:

### **Data Preprocessing**

#### **Data Collection & Cleaning**

```python
data = pd.read_csv("../datasets/Iris.csv")
data.head()
```

```python
data = data.drop(columns=["Id"])
data.head()
```

```python
data.info() # To check is there null values.
```

```python
data.describe() #
```

```python
x_vals = data.iloc[ : , 0:4]
x_vals[0:5]
```

```python
y_vals = data["Species"]
y_vals[0:5]
```

#### **Feature Scaling**

- Feature scaling is not required for tree based algorithms such as `Decision tree & Random Forest Algorithms`.

#### **Data Splitting**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_vals, y_vals)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### **Model Training**

- **DecisionTreeClassifier** Class used to *classification problems*

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X=x_train, y=y_train)
```

- **Prediction**

```python
y_pred = model.predict(x_test)
y_pred[0:10]
```

```python
y_test[0:10]
```

### **Model Evaluation**

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy = accuracy_score(y_true = y_test, y_pred = y_pred)
accuracy
```

```python
cm = confusion_matrix(y_true=y_test, y_pred=y_pred)
cm
```

### **Visualise the Decision Tree**

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12, 8)) 
plot_tree(model, feature_names=data.columns[0:-1], class_names=data["Species"], filled=True)
plt.show()
```
